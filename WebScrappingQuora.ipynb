```python
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, \
    WebDriverException
import time
import csv
import pandas as pd

flag = ''
num = 0
final_list = []
driver = webdriver.Chrome()

#Defining the function to scrape data
def scrape_quora(flag):
    
    driver.get("https://www.quora.com/topic/%s"% flag)
    for i in range(10):
        driver.execute_script("window.scrollBy(0,2000)", "")
        time.sleep(5)
    generateList()
            
#Defining the function to generate a list            
def generateList():
    
    element_list = []
    class_name = 'story_title_container'
    global final_list    
    try:
        all_elements = driver.find_elements_by_class_name(class_name)
        element_list = [x.text for x in all_elements]
        final_list = final_list + element_list
        
    except (NoSuchElementException, WebDriverException) as e:
        print(e)

#No. and pages to be scraped
while(num<=3):

    
    if(num == 0):
        scrape_quora(flag ='Mutual-Funds')
        
    elif(num == 1):

        scrape_quora(flag ='Indian-Mutual-Funds-1')
        
    elif(num == 2):
        
        scrape_quora(flag ='Mutual-Fund-Investment-Strategies')
        
    elif(num == 3):
        scrape_quora(flag ='Systematic-Investment-Plan-SIP')
    
    num = num + 1



print (final_list)
print('\n\n',len(final_list))   
    
#Getting the Data into CSV format

filename = "Quora_mutual_funds.csv"
  
# writing to csv file 
with open(filename, 'w', encoding="utf-8") as csvfile: 
    # creating a csv writer object 
    csvwriter = csv.writer(csvfile) 
    rows = zip(final_list)
    # writing the fields 
    csvwriter.writerow("Mutual funds Questions") 
      
    # writing the data rows 
    csvwriter.writerows(rows)

    
```
